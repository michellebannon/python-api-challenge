```python
from requests import get
import requests
```

```python
def getRhymes(word):
    return get('https://api.datamuse.com/words?rel_rhy=' + word).json()

wordToRhyme = 'Ramsey'

for rhyme in getRhymes(wordToRhyme):
    print(rhyme)
```
